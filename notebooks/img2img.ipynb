```python
import torch
from diffusers import FluxImg2ImgPipeline
pipe = FluxImg2ImgPipeline.from_pretrained("black-forest-labs/FLUX.1-dev", torch_dtype=torch.bfloat16)
```

```python
pipe.load_lora_weights('dolphinium/FLUX.1-dev-wikiart-impressionism', weight_name='flux_1_dev_wikiart_impressionism_v2.safetensors')
```

```python
device = "cuda" if torch.cuda.is_available() else "cpu"
pipe = pipe.to(device)
```

```python
from diffusers.utils import load_image
init_image = load_image("/content/image.jpg").resize((1024, 1024))
```

```python
prompt = "an impressionist landscape painting"
```

```python
image = pipe(
    prompt=prompt,
    image=init_image,
    num_inference_steps=20,
    strength=0.8,
    guidance_scale=9.0
).images[0]
```

```python
image.save("output_3.png")
```
